```
# Gr치fico 01

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.001)
y = -((x**2) - (2.5*x))

print(len(y))

f,c = plt.subplots()
c.plot(x,y)


plt.show()
```

```
# Gr치fico 02

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.001)
y = ((x**2)-1) / (x-1)

print(len(y))

f,c = plt.subplots()
c.plot(x,y, '--')


plt.show()
```

```
# Gr치fico 04

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.001)
y = []
x2 = np.arange(10., -10., -0.001)

for v in x:
  result = 0   # guaradar o valor da equacao escolhida
  
  if v <= -2:
    result = (-2*v)-1
    
  if -2 < v and v < 1:
    result = 3
    
  if 1 <= v:
    result = (2*v)+1
   
  y.append(result)
  
f,c = plt.subplots()
c.plot(x,y,x2,y, 'r')

c.grid()

plt.show()

print('..|End Application')
print( 'By Christopher')
```

```
import numpy as np
import matplotlib.pyplot as plt

def f01(x):
    return ((-2*x)-1)
  
def f02(x):
    return (2*x)+1 
  
def f00(x):
    if x <= -2:
         return f01(x)
      
    if -2 < x and x < 1:
         return 3
      
    if 1 <= x:
         return f02(x)
      
print ("Ola Christopher")      
x = np.arange (-10.,10.,0.001)  
y =[]

for v in x:
  y.append (f00(v))
  
fig, ax = plt.subplots()
ax.plot(x, y, '--''r')

ax.grid()
  
print(len(x))
print(len(y))
```

```
nome = []
salario = []
taxa = []

def printa_valores():
  for p, n in enumerate (nome):
    print ("{} - {} - {}".format(n, salario[p], taxa[p]))

  print('....')

def calcula_taxa (salario):
  if salario <= 1000:
    return 0
  elif salario > 1000 and salario <= 3000:
    return salario *0.15
  else:
    return salario *0.20
  
while True:
  v_nome = input ('Informe o Nome .: ')
  if v_nome.upper() == "FIM":
     break
  v_salario = float (input('Informe o Salario .: '))
  nome.append(v_nome)
  salario.append (v_salario)
  taxa.append (calcula_taxa(v_salario))
  
print (nome)
print (salario)
print (taxa)

print ('|..End Application')


```

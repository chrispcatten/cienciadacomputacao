```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

nome    = []
veiculo = []
valor   = [] 
ipva    = []
seguro  = []
v_seguro= 0
v_ipva  = 0
  
def calcula_1 (v_ipva): 
  return  v_valor * 0.04
def calcula_2 (v_seguro): 
  return v_valor * 0.0235
    
while True:
  v_nome = input ('Informe o Nome do Proprietário .: ')
  if v_nome.upper() == "FIM":
    break
  v_veiculo = input ('Informe o Nome do veiculo .: ') 

  v_valor = float (input('Informe o Valor do Veículo .: '))
  print('-------------------------------------------------')
  nome.append   (v_nome)
  veiculo.append(v_veiculo)
  valor.append  (v_valor)
  ipva.append   (calcula_1(v_ipva))
  seguro.append (calcula_2(v_seguro))

 
ind = np.arange(len(ipva))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, ipva,   width,   label='Ipva')
rects2 = ax.bar(ind + width/2, seguro, width,   label='Seguro')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel(' Valores Ipva / Seguro')
ax.set_xlabel(' Nomes dos carros')
ax.set_title('Tabela de Ipva e seguro')
ax.set_xticks(ind)
ax.set_xticklabels(veiculo)
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')

autolabel(rects1, "left")
autolabel(rects2, "right")

fig.tight_layout()

plt.show()

print (nome)
print (veiculo)
print (valor)
print (ipva)
print (seguro)

print ('|..End Application')
print ('By Christopher and Gabriel')
```

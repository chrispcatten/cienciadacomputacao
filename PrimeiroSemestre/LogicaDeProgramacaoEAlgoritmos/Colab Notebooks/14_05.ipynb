```
# EXERCIO 01

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.arange(-4.0, 4.0, 0.001)
y = (1)/1+(np.exp(-x))
x2  = np.arange(4.0, -4.0,- 0.001)

fig, ax = plt.subplots()
ax.plot(x, y, x2, y,'r')

ax.set(xlabel='escala X', ylabel='escala y',
       title='Exerc√≠cio 01')
ax.grid()

fig.savefig("grafico-01.png")
plt.show()

print ('|End Application')
```

```
#EXERCICIO 02

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.arange(-6.0, 8.0, 0.0001)
x1 = np.arange(6.0, -8.0, -0.0001)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

fig, ax = plt.subplots()
ax.plot(x, y,x1, y, 'r')

ax.set(xlabel='escala X', ylabel='escala y',
       title='Exerc√≠cio 02')
ax.grid()

fig.savefig("grafico-02.png")
plt.show()

print ('|End Application')

```
